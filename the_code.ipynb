Next, we will do a cross-correlation of all of the variables with each other to see if we can find any interesting correlations that should be analysed further.

```python
# Make a correlation matrix with the variables mentioned above
selected_columns = raw_data[["O_dlots", "O_Nlot", "O_aval", "O_daval", "R_bids", "R_dbids", "O_scope", "E_Pref", "R_pcost", "E_indice", "E_GDP_1"]]

# Have to make sure that the variables mentioned above are all floats (rather than integers)!
selected_columns = selected_columns.astype(float)

correlation_matrix = selected_columns.corr()

# create the heatmap of the correlation matrix
plt.figure(figsize = (8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt = ".2f")
plt.title("Correlation Matrix of selected columns")
plt.show()
```

From above, we can see that nothing is generated for the O_daval and the R_dbids variables. Let's inspect these further

```python
# check how many unique values the above mentioned variables contain:

print("O_daval has", raw_data["O_daval"].unique(), "unique variables")
print("R_dbids has", raw_data["R_dbids"].unique(), "unique variables")
```

We find that both variables only have one value! This means that for every process a bank deposit was required and a bid was always received. Thus, we can exclude them since they do not add any useful information to the data set. 

Looking at the rest of the correlation matrix, we see that the variables E_Pref and O_aval are perfectly correlated. This means that the values move in the same direction and at the same rate. This makes sense, since the deposit (O_aval) is dependant on the engineered reference price (E_Pref). As such, we only need to use one of the variables in our analysis. We will select O_aval, however either variable would've been suitable. 

We can also observe that R_pcost is almost perfectly correlated with E_Pref and O_aval (it has a score of 0.95). Thus the three variables move almost identically between themselves. Following the same explanation given above, we can also eliminate the R_pcost variable and only keep the O_aval variable in our analysis. 

From the rest of the data in correlation matrix, we do not see any more relations that are worthwhile of mention. As such, we can assume that the rest of the variables are not too closely correlated and thus fine to use.

### Competitive Analysis of the Data
This code segment categorizes rows  into different competitiveness levels (low, medium, and high) based on the distribution of bid counts in the "R_bids" column. It assumes that the data is roughly evenly distributed among these categories, utilizing quartiles to define the competitiveness ranges.


```python
# Find the quartiles of the R_bids variable
quartiles = raw_data["R_bids"].quantile([0, 0.25, 0.5, 0.75])

# Define the range boundaries based on quartiles
low_range = (float('-inf'), quartiles[0.25])
medium_range = (quartiles[0.25], quartiles[0.75])
high_range = (quartiles[0.75], float('inf'))

range_labels = ['Low', 'Medium', 'High']
raw_data['RangeCategory'] = pd.cut(raw_data['R_bids'], bins=[*low_range, *medium_range, *high_range], labels=range_labels, duplicates='drop')
nan_count = raw_data['R_bids'].isna().sum()

# Print the DataFrame with the assigned range categories and NaN count
print(f"Number of NaN values in 'R_bids' column: {nan_count} out of {len(raw_data)}")
print(f"Low range of competitiveness is {low_range}, meaning anything below {low_range[1]} bids is low")
print(f"Medium range of competitiveness is {medium_range}, meaning abything betwwen {medium_range[0]} and {medium_range[1]} bids is medium ")
print(f"High range of competitiveness is {high_range}, meaning anything below {high_range[0]} bids is high")
```

### Data Transformation

We will first of all remove the variables deemed as unnecessary in the Data Analysis exploraiton section. 

Additionally, we will remove the 17 rows of data that were found to have no data for the R_bids category (as seen in the Competitive Analysis of the Data setion) because this is a crucial variable for the construction of our model, and in order to ensure the integrity of our data we believe it is better to remove these processes altogether rather than estimate values for them. 

Next, we will check for duplicates within the remaining data. 

Then, we will check for outliers.

Finally, we will go through the data points of each variable to see if there are any points that logically cannot be correct (e.g. A country having a GDP = 0). 

```python
# Remove unnecessary variables from the data frame

cleaned_data_1 = raw_data.drop(columns = ["O_daval", "R_dbids", "E_Pref", "R_pcost", "O_countryp_code"])

cleaned_data_1
```

```python
# Next, remove the rows with NaN for the R_bids variables. 

cleaned_data_2 = cleaned_data_1.dropna(subset = "R_bids")
cleaned_data_2
```

```python
# Check for and remove any duplicate rows. 

#Note that we will only check if the entire row is duplicated rather than checking if the vast majority of variables are the same.
# This is because it is impossible to know that there aren't identical orders occuring in the same year, and because each contract
# has multiple entires it's impossible to use uniqueness of contract identifier as a sign of duplicates. 

duplicated_rows = cleaned_data_2.duplicated()

if duplicated_rows.any():
    print("There are duplicated rows in the Data set.")
else:
    print("There are no duplicated rows in the Data set.")


cleaned_data_3 = cleaned_data_2.drop_duplicates()
```

```python
# check for outliers within the data. Make a scatterplot of the remaining numerical values in the data set that have not yet been analysed or
# that are not identifiers (e.g. O_order, O_code, etc.) by year so as to get a visual overview of the data. 

years = cleaned_data_3["O_year_p"].values

# scatterplot for O_aval
O_aval = cleaned_data_3["O_aval"].values
plt.scatter(years, O_aval)
plt.title("Amount to be deposited as a guarantee to participate in the process")
plt.ylabel("Amount deposited (O_aval)")
plt.xlabel("Years")
plt.show()

# scatterplot for R_bids
R_bids = cleaned_data_3["R_bids"].values
plt.scatter(years, R_bids)
plt.title("Number of bids received from different bidders")
plt.ylabel("Number of bids received (R_bids)")
plt.xlabel("Years")
plt.show()

# scatterplot for E_indice
E_indice = cleaned_data_3["E_indice"].values
plt.scatter(years, E_indice)
plt.title("Engineered index relating the winning price to the engineered reference price")
plt.ylabel("Engineered index (E_indice)")
plt.xlabel("Years")
plt.show()

# scatterplot for E_GDP_1
E_GDP_1 = cleaned_data_3["E_GDP_1"].values
plt.scatter(years, E_GDP_1)
plt.title("Engineered value of the country GDP")
plt.ylabel("Engineered GDP (E_GDP_1)")
plt.xlabel("Years")
plt.show()

# scatterplot for E_DTF_1
E_DTF_1 = cleaned_data_3["E_DTF_1"].values
plt.scatter(years, E_DTF_1)
plt.title("Engineered variable for the openness of the country to foreign trade")
plt.ylabel("Engineered openness variable (E_DTF_1)")
plt.xlabel("Years")
plt.show()

# scatterplot for E_distance
E_distance = cleaned_data_3["E_distance"].values
plt.scatter(years, E_distance)
plt.title("Engineered average distance to the point of delivery of the lot")
plt.ylabel("Engineered average distance (E_distance)")
plt.xlabel("Years")
plt.show()

```

From the scatterplots above, we can see that all of the variables except for E_distance have some data points that are spread out far enough from the rest of the data that they could be outliers. In order to verify if this is the case (and thus remove the points), we will make a box plot for each variable and remove any points that are statistically considered outliers. 

```python
# boxplot for O_aval
plt.boxplot(cleaned_data_3["O_aval"])
plt.title("Amount to be deposited as a guarantee to participate in the process")
plt.show()

# boxplot for R_bids
plt.boxplot(cleaned_data_3["R_bids"])
plt.title("Number of bids received from different bidders")
plt.show()

# boxplot for E_indice
plt.boxplot(cleaned_data_3["E_indice"])
plt.title("Engineered index relating the winning price to the engineered reference price")
plt.show()

# boxplot for E_GDP_1
plt.boxplot(cleaned_data_3["E_GDP_1"])
plt.title("Engineered value of the country GDP")
plt.show()

# boxplot for E_DTF_1
plt.boxplot(cleaned_data_3["E_DTF_1"])
plt.title("Engineered variable for the openness of the country to foreign trade")
plt.show()

```

Looking at the boxplots above, we can see that there are a large number of rows that qualify as outliers between all of the variables. However, we do not want to risk removing too many data points, especially when it's not necessarily needed, as that would misrepresent the data. Looking at the ranges of the various variables' data, we can also logically deduce that for all of the variables (except for O_aval) the spread is actually not too dispersed. O_aval, instead, is the only variable with a very noticable and big spread (the vast majority of data points have a value of less than 100,000 but there are some points reaching values as high as 600,000). As such, we decided to only use O_aval to determine outliers within the data set. 



```python
# Calculate and remove the outliers based off of the O_aval variable 

Q1 = np.percentile(cleaned_data_3["O_aval"], 25)
Q3 = np.percentile(cleaned_data_3["O_aval"], 75)
IQR = Q3-Q1
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR

cleaned_data_4 = cleaned_data_3[(cleaned_data_3["O_aval"]>= lower_bound) & (cleaned_data_3["O_aval"]<= upper_bound)]
 
 
# remake the boxplots 

# boxplot for O_aval
plt.boxplot(cleaned_data_4["O_aval"])
plt.title("Amount to be deposited as a guarantee to participate in the process")
plt.show()

# boxplot for R_bids
plt.boxplot(cleaned_data_4["R_bids"])
plt.title("Number of bids received from different bidders")
plt.show()

# boxplot for E_indice
plt.boxplot(cleaned_data_4["E_indice"])
plt.title("Engineered index relating the winning price to the engineered reference price")
plt.show()

# boxplot for E_GDP_1
plt.boxplot(cleaned_data_4["E_GDP_1"])
plt.title("Engineered value of the country GDP")
plt.show()

# boxplot for E_DTF_1
plt.boxplot(cleaned_data_4["E_DTF_1"])
plt.title("Engineered variable for the openness of the country to foreign trade")
plt.show()

```
