## Competitive Analysis of the Data
This code segment categorizes rows  into different competitiveness levels (low, medium, and high) based on the distribution of bid counts in the "R_bids" column. It assumes that the data is roughly evenly distributed among these categories, utilizing quartiles to define the competitiveness ranges.


```python
# Find the quartiles of the R_bids variable
quartiles = raw_data["R_bids"].quantile([0, 0.25, 0.5, 0.75])

# Define the range boundaries based on quartiles
low_range = (float('-inf'), quartiles[0.25])
medium_range = (quartiles[0.25], quartiles[0.75])
high_range = (quartiles[0.75], float('inf'))

range_labels = ['Low', 'Medium', 'High']
raw_data['RangeCategory'] = pd.cut(raw_data['R_bids'], bins=[*low_range, *medium_range, *high_range], labels=range_labels, duplicates='drop')
nan_count = raw_data['R_bids'].isna().sum()

# Print the DataFrame with the assigned range categories and NaN count
print(f"Number of NaN values in 'R_bids' column: {nan_count} out of {len(raw_data)}")
print(f"Low range of competitiveness is {low_range}, meaning anything below {low_range[1]} bids is low")
print(f"Medium range of competitiveness is {medium_range}, meaning abything betwwen {medium_range[0]} and {medium_range[1]} bids is medium ")
print(f"High range of competitiveness is {high_range}, meaning anything above {high_range[0]} bids is high")
```

We will use these ranges (generated before cleaning the data set) to find the range of bids aligned with each competitive level. This is because, even though the model will be generated using a cleaned data set, we want to keep the original data as the marker for competitivness as it is more representative of reality and real-life processes.

## Machine Learning Model

### Choosing classification techniques to test

**Linear regression**

linear regression is a statistical technique used to study the relation among variables. Since our data set isn’t that large, we find that linear regression is one of the machine learning models we should consider. It is one of the simplest models and it provides clear and interpretable results. Moreover, linear regression is well-suited for modeling relationships between continious variables. Some of the variables in our data set (such as “E_Pref”, “E_indice”, “E_GDP_1” and “E_DTf_1”) are continous variables. Thus, since it can effectively capture linear relationships between continous variables, linear regression will be one of the two models we will train and test. Our goal is to predict the suitability or unsuitability of procurement opportunities based on the variables we have, and linear regression will provide us with coefficienct for each of those variables, helping us pinpoint which factors most affect the procurement suitability.



**Decision trees**

Another one of the models possible to use for this Machine Learning problem is decision trees. This model is extremely efficient, which is an important advantage as we are working with a relatively small dataset. Furthermore, decision trees are advantageous due to their interpretability, which is important for this Machine Learning problem as it allows WANTWIN to understand the features of procurement processes that are important for classifying viability. Decision trees are easy to visualize which makes it a valuable tool for explaining the decision-making process to stakeholders. Although decision trees could be prone to overfitting, this is a problem we can subvert by the selection of appropriate hyperparameters. Decision trees are most efficient at solving interperatable classification problems and the element of feature selection is extremely useful to provide relevant infromation to WANTWIN.


**Naïve Bayes classification**

Albeit being a simple classification technique to use, the Naïve Bayes classification is only really suitable for simple, independent data sets. This is due to a fundamental assumption of this classification technique: the Assumption of Independence. For the Naïve Bayers classification to provide optimal and accurate results, we need features to be conditionally independent between themselves. In most real-life scenarios, however, this does not hold as features are very often correlated. The data set being used for this model is comes from real-life data. In addition, it has 17 variables, meaning that the likelihood of the features being independent is almost zero. As such, using the Naïve Bayes classification would lead to a model that would not be accurate; thus, we will not consider this technique for our model. 

**K-nearest neighbours (KNN)**

KNN is a classification technique that is very easy to implement and simple to interpret, however its main drawback is that it can only handle simple data sets (in other words, datasets with few dimensions). When the number of dimensions is high, an extremely high amount of data is required to maintain a good performance of the model. In addition, the distance metric used to measure the similiarity between the data points becomes less meaningful; this is due to the "crowding problem" (which is when, in high dimensions, data points appeard to be equally distant from one another). Considering that the data set being used to train this model has 17 categories (dimensions) but only 337 rows, it is safe to assume that this is simply too many dimensions for KNN to be efficient and produce a high-quality model. As such, we will not consider this classification technique for our model.

We have calculated that the best parameters to use are 8 for 'max_depth' and 5 for 'min_samples_split', giving us an accuracy of on average 75% (keep in mind that the accuracy changes every time the model is run).

```python
from sklearn.model_selection import GridSearchCV

# Define the parameters you want to tune
param_grid = {
    'max_depth': [2,4,6,8],  # Define a standard range of maximum depth values
    'min_samples_split': [3,5,7,9]  # Define a standard range of minimum samples split
}

# Initialize the DecisionTreeClassifier
clf_2 = DecisionTreeClassifier()

# Perform grid search with cross-validation
grid_search = GridSearchCV(clf_2, param_grid, cv=5)
grid_search.fit(x_train, y_train)  

# Get the best parameters
best_clf = grid_search.best_estimator_

# Make predictions on the test data using the best model
y_pred = best_clf.predict(x_test)  

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Best Parameters:", grid_search.best_params_)
```
